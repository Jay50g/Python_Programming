```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
r = np.linspace(2.5, 4.0, 500) 
n = 1000  
last = 100 
```

```python
x = 1e-5 
X = np.zeros((n,len(r)))
```

```python
for i in range(n):
    x = r * x * (1 - x)
    X[i]=x.copy()
```

```python
X.shape
```

```python
plt.plot([r]*last,X[-last:,:], 'r.',markersize=1)
plt.xlabel('Growth Rate')
plt.ylabel('Population')
p1=plt.title('Logistic Map Bifurcation')
plt.grid(alpha=0.4)
```

```python
plt.plot([r]*last,X[:last,:], 'r.',markersize=1)
plt.xlabel('Growth Rate')
plt.ylabel('Population')
p1=plt.title('Logistic Map Bifurcation')
plt.grid(alpha=0.4)
```
