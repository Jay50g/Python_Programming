```python
# move the first odd number to the back of the list

a= [10,2,4,7,6,7,8,9]
```

```python
def pushOddBack(x):

    for i in range(len(x)):
        if x[i]%2==1:
            x.append(x.pop(i))
            break
```

```python
# record the occurrences of characters in a string
def freq(x):
    f={}
    for i in x:
      if i in f:
        f[i]+=1
      else: f[i]=1
    return f
```

```python
print(freq(a))
```

```python
# replace a group of characters with others
# Caeser cipher

# encryption code
replacement={'l':'b', 'e':'a','y':'c','i':'t'}
```

```python
def replace(x,y):
    out=x
    for i in out:
        if i in y:
            out=out.replace(i,y[i])
    return out
```

```python
replace('Hello, My friend!', replacement)
```

```python
# invert a dictionary, exchange the index with its value
# build the decryption codes

def invtDic(x):
    f={}
    for i in x:
        f[x[i]]=i
    return f
```

```python
# decryption code
invtDic(replacement)
```

```python
# encrypt and decrypt the message.

replace(replace('Hello, My friend!', replacement),\
          invtDic(replacement))
```
