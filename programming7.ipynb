```python
# This file shows functions and recursions

# Calculate factorial with recursion

def fac(x):
    if x==0: return 1
    else: return x*fac(x-1)

print(fac(5))
```

```python
# C curves

import turtle as tt
tt.color('red')
tt.pensize(3)
tt.speed(0)

def drawC(length,level):
    tt.down()
    if level>1:
        tt.lt(45)
        drawC(length/2,level-1)
        tt.rt(90)
        drawC(length / 2, level - 1)
        tt.lt(45)
    else: tt.fd(length)

drawC(4000,10)
```
