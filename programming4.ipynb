```python
# draw a stop sign

import turtle as tt
tt.up()
tt.goto(-100,100)
tt.color('red')
tt.down()
tt.begin_fill()
for i in range(6):
    tt.fd(200)
    tt.rt(60)
tt.end_fill()
tt.up()
tt.goto(-145,-140)
tt.color('white')
tt.ht()
tt.write('STOP',font=('Arial',80,'normal'))
tt.done()
```
