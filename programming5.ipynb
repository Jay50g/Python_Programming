```python
# for loops , while loops,
# normal distribution with turtle
# Monte Carlo method to calculate pi
# Guess a number within 1000


# Calculate Pi with Monte Carlo

import random as rdm
import turtle as tt
import math as mt
t1=tt.Turtle()
t2=tt.Turtle()
t3=tt.Turtle()
t3.color('red')
t2.up()
t3.up()
t2.ht()
t3.ht()
t2.goto(-600,240)
t2.color('green')
t2.write('Serial Monte Carlo to Calculate PI',\
         font=('Arial',40,'normal'))


t1.color('red')
t1.pensize(5)
t1.up()
t1.goto(0,-200)
t1.down()
t1.circle(200)
t1.up()
rdm.seed(3)
inside=0
k=0
for i in range(1000):
    x=rdm.uniform(-200,200)
    y=rdm.uniform(-200,200)
    if mt.sqrt(x**2+y**2)<200:
        inside+=1
        PI=4*inside/(i+1)
    t1.goto(x,y)
    t1.dot(5,'blue')
    t3.goto(-500,200-k)
    t3.write(str(format(round(PI,4),'5.4f')), font=('Arial',38, 'normal'))
    k+=50
    if (i+1)%12==0:
        k=0
        t3.clear()
```

```python
# Normal Distribution

import turtle as tt
tt.up()
tt.goto(-600,240)
tt.color('green')
tt.write('Scientific Computing  (Simple Random Sampling)',\
         font=('Arial',40,'normal'))
tt.speed(0)
tt.color('blue')
p=[0]*500
import random as rdm
for i in range(100000):
  x=int(rdm.gauss(0,2)/0.1)*4
  #print(x)
  p[x]+=0.5
  if i%30==0:
    tt.goto(x,p[x]-300)
    tt.dot(5,'blue')
```

```python
# Guess a number within 1000

import random as rdm
x=rdm.randint(0,1001)
times=0
print('The system has generated a number within[0,1000]')
while 1:
    y=eval(input('Please input your guessed number!'))
    times+=1
    print('You have guessed ',times,' times!')
    if y==x:
        print('Excellent, you are right!')
        break
    elif y>x :print('Too big!')
    else: print('Too small!')
```
