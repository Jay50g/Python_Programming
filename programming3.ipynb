```python
import turtle as tt
t1=tt.Turtle()
t1.up()                  # away from the paper
t1.down()                # pen on the paper
t1.pensize(5)            # pen width
t1.fd(100)               # move forward 100 pixels
t1.bk(100)               # move backward 100 pixels
t1.heading()             # show direction
t1.seth(90)              # set direction
t1.lt(30)       # t1.left(30), left 30
t1.rt(30)       # t1.right(30), right 30
t1.color('red')          # set pen color
```

```python
# draw polygon radius,degrees,sides
t1.circle(100,180,100)
t1.speed(2)             # set speed
t1.degrees()            # set to degrees
t1.radians()            # set to radians
t1.st()                 # show turtle
t1.ht()                 # pen invisible
t1.begin_poly()         # record vertex(x,y)
t1.get_poly()           # read vertex(x,y)
t1.filling()            # True fill, False no fill
```

```python
# draw a red filled circle
t1.begin_fill()
t1.fillcolor('red')
t1.circle(100)
t1.end_fill()
```

```python
# return the distance from current position to (x,y)
t1.distance(100,20)
t1.pos()            # return the current position
```

```python
# draw a filled blue circle with diameter 100,
t1.dot(100,'blue')
```

```python
t1.home()  # goto(0,0)
t1.goto(200,30) # goto (200,30)
```

```python
# write words on the paper
t1.color('green')
t1.write('hello', font=('Arial', 40, 'normal'))
```
