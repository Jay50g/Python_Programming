```python
class Book:
    numberOfBooks=0   # class static variable
    genres={}
    # object constructor
    def __init__(self,genre,title, author, py):
        self.genre=genre
        self.title=title
        self.author=author
        self.publishYear=py
        Book.numberOfBooks+=1

    def calculateAge(self, year):
        return year-self.publishYear

    def addToGenres(x):
        if x.genre in Book.genres:
            Book.genres[x.genre].append([x.title, x.author])
        else:
            Book.genres[x.genre]=[[x.title,x.author]]
```

```python
# this program is very difficult. It is not for beginners
# hello ----  ellohay

def pig_latin(x):
    vow=['a','e','i','o','u']
    table = [chr(x) for x in range(ord('A'), ord('Z') + 1)]
    table = ''.join(table)
    table += table.lower()
    a=x.split()   # split the string into a word list
    for i in range(len(a)):
        # if there is a punctuation mare behind of a word
        if a[i][-1] not in table:
            # remover the Punctuation Mark behind of a word
            y=a[i][0:-1]
            y2=y
            for j in range(len(y)):
               if y[j] in vow:
                  y=y[j:]+y[0:j] +'ay'
                  break
            # if there is no vowel in the word
            if y2==y and len(a[i])!=1 : y+='ay'
            # put the punctuation mark back
            a[i]=y+a[i][-1]
        else:  # in case there is no punctuation mark
            y = a[i]
            y2 = y
            for j in range(len(y)):
                if y[j] in vow:
                    y = y[j:] + y[0:j] + 'ay'
                    break
            if y2 == y: y += 'ay'
            a[i] = y

    return ' '.join(a)
```

```python
# count the occurrence of words in a text

def count_words(x):
    f={}
    y=x.split()
    for i in y:
        if i not in f:
            f[i]=1
        else: f[i]+=1
    return f
```

```python
# Natural language processing with a Markov Chain

# cut k characters from the text to build the kgram
def get_kgram(text, pos, k):
    return text[pos:pos+k]
```

```python
markovModel={}
```

```python
#
def process_character(m,text,pos,k):
    f={}
    kg=get_kgram(text,pos,k)
    if kg not in m:         # new kgram
        f[text[pos+k]]=1
        m[kg] =f    # build kg as a new key and assign f to it
    # kgram is there, but the followed character is not there
    elif text[pos+k] not in m[kg]:
        f=m[kg]    # save the old characters and their frq
        f[text[pos + k]] = 1   # and the new character and its frq
        m[kg] = f
    else:     # if the kgram and the followed character is there
        f = m[kg]  # save the old characters and their frq
        f[text[pos + k]] += 1    # just add one more time to the frq
        m[kg] = f
```

```python
''' # To test the function if it can add keys and values 
        to the markovModel. 

process_character(markovModel,'the ether',0,4)
process_character(markovModel,'the ether',1,4)
process_character(markovModel,'the ether',2,4)
'''

def build_markovModel(text,k):
    for i in range(len(text)-k):
        process_character(markovModel,text,i,k)
    text=text[len(text)-k:]+text[0:k]
    for i in range(k):
        process_character(markovModel,text,i,k)
```

```python
import random as rdm
```

```python
def random_character(m,kgram):
    base=''
    keys=list(m[kgram].keys())
    values=list(m[kgram].values())
    for i in range(len(keys)):
        base+=keys[i]*values[i]
    return rdm.choice(base)
```

```python
def generate_random_text(m,N):
    seed=rdm.sample(list(m.keys()),k=1)
    seed=''.join(seed)
    text=''
    for i in range(N):
        x=random_character(m,seed)
        seed=seed[1:]+x
        text+=x
    return text
```
