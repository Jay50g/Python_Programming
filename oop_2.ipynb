```python
# =============================================================================
# BASE CLASS (PARENT CLASS)
# =============================================================================
class Vehicle:
    vehicle_count=0

    
    def __init__(self, brand, model, year, color):
        """
        Constructor method - called automatically when creating a new object
        'self' refers to the current instance being created
        """
        
        # Instance attributes - unique to EACH object
        self.brand = brand      # Public attribute
        self.model = model      # Public attribute  
        self.year = year        # Public attribute
        self.color = color      # Public attribute
        self._mileage = 0       # Protected attribute (convention: single underscore)
        self.__vin = self._generate_vin()  # Private attribute (double underscore)
        
        # Update class attribute
        Vehicle.vehicle_count += 1
        
        print(f"‚úÖ Vehicle object created: {brand} {model}")
        print(f"   Total vehicles created: {Vehicle.vehicle_count}")
    
    def _generate_vin(self):
        """Protected method - for internal use only"""
        import random
        return f"VIN{random.randint(10000, 99999)}"
    
    # INSTANCE METHODS
    def start_engine(self):
        """Instance method - operates on a specific vehicle object"""
        print(f"üöó {self.brand} {self.model}'s engine is starting...")
        return True
    
    def drive(self, distance):
        """Instance method that modifies object state"""
        self._mileage += distance
        print(f"üìç {self.brand} {self.model} drove {distance} miles. Total mileage: {self._mileage}")
        return self._mileage
    
    def get_info(self):
        """Public method to get vehicle information"""
        info = f"{self.year} {self.brand} {self.model} ({self.color})"
        return info
    
    def get_mileage(self):
        """Public method to access protected attribute"""
        return self._mileage
    
    def get_vin(self):
        """Public method to access private attribute (encapsulation)"""
        return self.__vin
    
    # CLASS METHOD
    @classmethod
    def get_vehicle_count(cls):
        """Class method - operates on the CLASS, not instances"""
        return cls.vehicle_count
    
    # STATIC METHOD  
    @staticmethod
    def is_vehicle_old(year):
        """Static method - doesn't need class or instance reference"""
        current_year = 2024
        return current_year - year > 10
    
    # SPECIAL METHOD (Magic/Dunder method)
    def __str__(self):
        """String representation of the object - called by print()"""
        return f"Vehicle: {self.get_info()}"


# =============================================================================
# DERIVED CLASS (CHILD CLASS) - INHERITS FROM VEHICLE
# =============================================================================
class Car(Vehicle):  # Car inherits from Vehicle
    """
    Child class that inherits from Vehicle.
    Car has all Vehicle attributes/methods PLUS its own.
    """
    
    def __init__(self, brand, model, year, color, doors, fuel_type):
        """
        Child class constructor
        First, we call the parent constructor using super()
        Then we initialize child-specific attributes
        """
        
        # Call the parent class constructor
        super().__init__(brand, model, year, color)
        
        # Child-specific attributes
        self.doors = doors
        self.fuel_type = fuel_type
        self.__radio_station = "FM 98.5"  # Private to Car class
        
        print(f"   üöô Car-specific: {doors} doors, {fuel_type} fuel")
    
    # Method overriding - same method name as parent, different implementation
    def start_engine(self):
        """Override parent method - Cars start differently"""
        print(f"üîë {self.brand} {self.model} car engine starting with key...")
        print(f"   Fuel type: {self.fuel_type}")
        return True
    
    # New method specific to Car class
    def honk_horn(self):
        """Method only available to Car objects"""
        print(f"üìØ {self.brand} {self.model} goes BEEP BEEP!")
    
    def change_radio(self, station):
        """Method to modify private attribute"""
        self.__radio_station = station
        print(f"üìª Radio changed to: {station}")
    
    def get_radio(self):
        """Method to access private attribute"""
        return self.__radio_station
    
    # Override the get_info method from parent
    def get_info(self):
        """Extend parent method with car-specific info"""
        base_info = super().get_info()  # Call parent method
        car_info = f"{base_info} | {self.doors} doors | {self.fuel_type}"
        return car_info


# =============================================================================
# ANOTHER CHILD CLASS
# =============================================================================
class Motorcycle(Vehicle):
    """Another child class inheriting from Vehicle"""
    
    def __init__(self, brand, model, year, color, engine_size):
        super().__init__(brand, model, year, color)
        self.engine_size = engine_size  # in cc
    
    # Method overriding
    def start_engine(self):
        """Motorcycles start differently"""
        print(f"üèçÔ∏è {self.brand} {self.model} motorcycle engine roaring to life!")
        return True
    
    def wheelie(self):
        """Method only for Motorcycle objects"""
        print(f"üéØ {self.brand} {self.model} is doing a wheelie! üöÄ")


# =============================================================================
# DEMONSTRATION AND EXPLANATION
# =============================================================================
def demonstrate_oop_concepts():
    print("=" * 70)
    print("OBJECT-ORIENTED PROGRAMMING DEMONSTRATION")
    print("=" * 70)
    
    # =========================================================================
    # 1. CREATING OBJECTS (INSTANTIATION)
    # =========================================================================
    print("\n1. CREATING OBJECTS (INSTANTIATION)")
    print("-" * 40)
    
    # Create a Vehicle object
    print("Creating Vehicle object:")
    vehicle1 = Vehicle("Toyota", "Generic", 2020, "Blue")
    # Syntax: object_name = ClassName(arguments)
    
    # Create a Car object (child class)
    print("\nCreating Car object:")
    car1 = Car("Honda", "Civic", 2022, "Red", 4, "Gasoline")
    # Car inherits from Vehicle, so it needs Vehicle arguments PLUS its own
    
    # Create a Motorcycle object (another child class)
    print("\nCreating Motorcycle object:")
    bike1 = Motorcycle("Yamaha", "R1", 2023, "Black", 1000)
    
    # =========================================================================
    # 2. ACCESSING ATTRIBUTES AND METHODS
    # =========================================================================
    print("\n\n2. ACCESSING ATTRIBUTES AND METHODS")
    print("-" * 40)
    
    # Access instance attributes
    print(f"Vehicle 1 brand: {vehicle1.brand}")        # Public attribute
    print(f"Car 1 doors: {car1.doors}")                # Child class attribute
    print(f"Motorcycle engine: {bike1.engine_size}cc") # Child class attribute
    
    # Access protected attribute through method
    print(f"Car 1 mileage: {car1.get_mileage()} miles")
    
    # Access private attribute through method
    print(f"Car 1 VIN: {car1.get_vin()}")
    
    # =========================================================================
    # 3. CALLING METHODS
    # =========================================================================
    print("\n\n3. CALLING METHODS")
    print("-" * 40)
    
    # Call instance methods
    print("Calling start_engine():")
    vehicle1.start_engine()  # Vehicle's version
    car1.start_engine()      # Car's overridden version  
    bike1.start_engine()     # Motorcycle's overridden version
    
    print("\nCalling drive():")
    car1.drive(50)   # Inherited from Vehicle
    bike1.drive(25)  # Inherited from Vehicle
    
    print("\nCalling child-specific methods:")
    car1.honk_horn()     # Only available to Car objects
    bike1.wheelie()      # Only available to Motorcycle objects
    
    # =========================================================================
    # 4. INHERITANCE DEMONSTRATION
    # =========================================================================
    print("\n\n4. INHERITANCE DEMONSTRATION")
    print("-" * 40)
    
    # Car and Motorcycle inherit all Vehicle methods
    print("Methods inherited from Vehicle:")
    print(f"Car info: {car1.get_info()}")        # Uses Car's overridden version
    print(f"Bike info: {bike1.get_info()}")      # Uses Vehicle's version
    
    # Demonstrate method overriding
    print("\nMethod Overriding:")
    car1.change_radio("Rock FM 101.1")
    print(f"Car radio: {car1.get_radio()}")
    
    # =========================================================================
    # 5. CLASS METHODS AND STATIC METHODS
    # =========================================================================
    print("\n\n5. CLASS AND STATIC METHODS")
    print("-" * 40)
    
    # Class method - called on the class, not instance
    print(f"Total vehicles created: {Vehicle.get_vehicle_count()}")
    
    # Static method - doesn't need object or class reference
    print(f"Is 2010 vehicle old? {Vehicle.is_vehicle_old(2010)}")
    print(f"Is 2022 vehicle old? {Vehicle.is_vehicle_old(2022)}")
    
    # =========================================================================
    # 6. POLYMORPHISM
    # =========================================================================
    print("\n\n6. POLYMORPHISM")
    print("-" * 40)
    
    # Different objects, same method name, different behaviors
    vehicles = [vehicle1, car1, bike1]
    
    for vehicle in vehicles:
        print(f"\n{vehicle.get_info()}:")
        vehicle.start_engine()  # Each calls its own version
    
    # =========================================================================
    # 7. ENCAPSULATION
    # =========================================================================
    print("\n\n7. ENCAPSULATION (DATA HIDING)")
    print("-" * 40)
    
    # Public attributes - directly accessible
    print(f"Public: Car color is {car1.color}")
    
    # Protected attributes - accessible but shouldn't be used directly
    print(f"Protected: Car mileage is {car1._mileage} (accessed directly - not recommended)")
    
    # Private attributes - name mangling makes them harder to access
    try:
        # This will fail - private attribute
        print(car1.__vin)
    except AttributeError as e:
        print(f"Private: Cannot access __vin directly: {e}")
    
    # Proper way to access private attributes
    print(f"Private: Car VIN via method: {car1.get_vin()}")


# =============================================================================
# ADVANCED EXAMPLE: USING THE CLASSES
# =============================================================================
def create_vehicle_fleet():
    """Create and manage a fleet of vehicles"""
    
    print("\n" + "=" * 70)
    print("VEHICLE FLEET MANAGEMENT")
    print("=" * 70)
    
    # Create multiple vehicles
    fleet = [
        Car("Toyota", "Camry", 2023, "Silver", 4, "Hybrid"),
        Car("Ford", "Mustang", 2022, "Blue", 2, "Gasoline"),
        Motorcycle("Harley", "Davidson", 2021, "Black", 1200),
        Car("Tesla", "Model 3", 2024, "White", 4, "Electric"),
        Vehicle("Generic", "Truck", 2020, "Yellow")
    ]
    
    # Fleet operations
    total_mileage = 0
    
    for i, vehicle in enumerate(fleet, 1):
        print(f"\n{i}. {vehicle.get_info()}")
        
        # Polymorphism in action
        vehicle.start_engine()
        
        # Simulate some driving
        miles = (i + 1) * 10
        vehicle.drive(miles)
        total_mileage += miles
        
        # Call type-specific methods
        if isinstance(vehicle, Car):
            vehicle.honk_horn()
        elif isinstance(vehicle, Motorcycle):
            vehicle.wheelie()
    
    print(f"\nüìä FLEET SUMMARY:")
    print(f"Total vehicles: {len(fleet)}")
    print(f"Total mileage driven: {total_mileage} miles")
    print(f"Vehicle count via class: {Vehicle.get_vehicle_count()}")


# =============================================================================
# MAIN EXECUTION
# =============================================================================
if __name__ == "__main__":
    # Run the demonstration
    demonstrate_oop_concepts()
    
    # Create a vehicle fleet
    create_vehicle_fleet()
    
    # =========================================================================
    # SYNTAX EXPLANATION SUMMARY
    # =========================================================================
    print("\n" + "=" * 70)
    print("KEY SYNTAX EXPLANATIONS")
    print("=" * 70)
    
    explanations = {
        "class Definition": "class ClassName: - defines a new class",
        "Inheritance": "class Child(Parent): - Child inherits from Parent",
        "__init__ method": "Constructor - called when creating new objects",
        "self parameter": "Refers to the current object instance", 
        "super()": "Calls parent class methods from child class",
        "@classmethod": "Method that operates on the class, not instances",
        "@staticmethod": "Method that doesn't need class/instance reference",
        "Public attribute": "self.attribute - accessible from anywhere",
        "Protected attribute": "self._attribute - internal use (convention)",
        "Private attribute": "self.__attribute - name mangling, harder to access"
    }
    
    for concept, explanation in explanations.items():
        print(f"‚Ä¢ {concept:20} ‚Üí {explanation}")
    
```
