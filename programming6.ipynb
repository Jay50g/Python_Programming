```python
# this file shows various applications of loops


# Display a table of characters and their Ascii codes
print('='*60)
print('    Characters and their ASCII codes ')
print('='*60)
numbers=ord('~')-ord('!')
k=0
while(k<numbers):
  print('ASCII   |   ',end='')
  for i in range(10):
    if i+k>numbers: break
    print(format(chr(i +k+ ord('!')), '<4s'), end=' ')
  print()
  print('DEC     |   ',end='')
  for i in range(10):
    if i + k > numbers: break
    print(format(i + k+ord('!'), '<4d'), end=' ')
  print()
  print('HEX     |   ',end='')
  for i in range(10):
    if i + k > numbers: break
    print(format(i + k+ord('!'), '<4x'), end=' ')
  print()
  print('=' * 60)
  k+=10
```

```python
# Display a pyramid of 2's exponents

for i in range(8):
  for j in range(7-i):
    print('   ',end=' ')
  for j in range(i+1):
    print(format(2**j,'3d'),end=' ')
  for j in range(i):
    print(format(2**(i-j-1),'3d'),end=' ')
  print()
```

```python
qPi=0
sn=1
for i in range(1,int(1e7),2):
  qPi+=1/(i*sn)
  sn*=-1
print('The calculated PI value is ',4*qPi)
```

```python
import turtle as tt
import math as mt
def drawAxis(howLong,centerXY):
  tt.up()
  tt.goto(centerXY[0],centerXY[1])
  tt.bk(howLong/2)
  tt.pensize(4)
  tt.down()
  tt.fd(howLong)
  tt.lt(150)
  tt.fd(30)
  tt.bk(30)
  tt.lt(60)
  tt.fd(30)
  tt.up()

tt.speed(5)
tt.seth(0)
drawAxis(1200,(0,0))


tt.seth(90)
drawAxis(600,(0,0))


def drawFun(fun):
  for x in range(-600,600,2):
    if x==-600:
      tt.goto(x,150*fun(x/60))
    tt.down()
    tt.goto(x, 150 * fun(x /60))
  tt.up()

tt.speed(0)
tt.color('red')
drawFun(mt.sin)
tt.color('blue')
drawFun(mt.cos)

tt.done()
```
